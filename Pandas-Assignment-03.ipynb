```python
# # Q2. What is the difference between the functions loc and iloc in pandas.DataFrame?
# import pandas as pd
#  course_name = [‘Data Science’, ‘Machine Learning’, ‘Big Data’, ‘Data Engineer’]
#  duration =  [2,3,6,4]
#  df = pd.DataFrame(data = {‘course_name’ : course_name, ‘duration’ : duration})
# Answer :-
# In pandas, the loc and iloc functions are used to access and manipulate data in a DataFrame. Here's the difference between the two:

# loc:

# loc is label-based indexing, which means it is used to access rows or columns by their labels.
# It takes two parameters separated by a comma: the row label(s) and column label(s).
# The labels can be either a single label, a list of labels, or a slice object.
# Example usage: df.loc[2, 'duration'] or df.loc[1:3, 'course_name'].
# iloc:

# iloc is integer-based indexing, which means it is used to access rows or columns by their integer positions.
# It takes two parameters separated by a comma: the row index(es) and column index(es).
# The indices can be either a single index, a list of indices, or a slice object.
# Example usage: df.iloc[2, 1] or df.iloc[1:3, 0].
# To summarize, loc is used when accessing data using labels (row or column names), while iloc is used when accessing data using integer positions (row or column indices).




```

```python
# Q6. What do you understand about the windows function in pandas and list the types of windows 
# functions.
# import pandas as pd
#  import numpy as np
#  columns = ['column_1', 'column_2', 'column_3', 'column_4', 'column_5', 'column_6']
#  indices = [1,2,3,4,5,6]
#  #Creating a dataframe:
#  df1 = pd.DataFrame(np.random.rand(6,6), columns = columns, index = indices)
# Answer :-
# In pandas, a window function is a way to perform calculations on a specific subset of data within a larger dataset. It allows you to define a window or a group of rows on which you can apply various functions. The window function calculates the result based on the values within the defined window.

# Pandas provides several types of window functions, including:

# Rolling Window Functions:

# Rolling window functions operate on a defined window of a specified size that "rolls" or moves along the data.
# Examples of rolling window functions include rolling mean (rolling().mean()), rolling sum (rolling().sum()), rolling standard deviation (rolling().std()), etc.
# These functions are useful for analyzing trends and patterns in time series data.
# Expanding Window Functions:

# Expanding window functions calculate the result based on all the previous values up to the current row.
# Examples of expanding window functions include cumulative sum (expanding().sum()), cumulative maximum (expanding().max()), cumulative minimum (expanding().min()), etc.
# These functions are helpful in analyzing cumulative or aggregate data over time.
# Exponential Weighted Functions:

# Exponential weighted functions assign weights to data points based on their proximity to the current point, giving more weight to recent data.
# Examples of exponential weighted functions include exponentially weighted moving average (ewm().mean()), exponentially weighted sum (ewm().sum()), etc.
# These functions are useful for calculating smoothed or weighted averages.
# Window functions are commonly used for time series analysis, data smoothing, trend analysis, and calculating rolling or cumulative metrics. They provide flexibility in analyzing and summarizing data within specific windows or groups.





```

```python
# Q8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) and 
# calculates the difference between them in days, hours, and minutes using Pandas time delta. The 
# program should prompt the user to enter the dates and display the result
# import pandas as pd
#  import numpy as np
#  columns = ['column_1', 'column_2', 'column_3', 'column_4', 'column_5', 'column_6']
#  indices = [1,2,3,4,5,6]
#  #Creating a dataframe:
#  df1 = pd.DataFrame(np.random.rand(6,6), columns = columns, index = indices)
# Answer :-
import pandas as pd

def get_time_difference(start_date, end_date):
    start_dt = pd.to_datetime(start_date)
    end_dt = pd.to_datetime(end_date)
    time_delta = end_dt - start_dt
    days_diff = time_delta.days
    hours_diff = time_delta.seconds // 3600
    minutes_diff = (time_delta.seconds // 60) % 60
    return days_diff, hours_diff, minutes_diff

def main():
    try:
        start_date = input("Enter the start date (YYYY-MM-DD): ")
        end_date = input("Enter the end date (YYYY-MM-DD): ")
        days_diff, hours_diff, minutes_diff = get_time_difference(start_date, end_date)
        print("Difference:")
        print("Days:", days_diff)
        print("Hours:", hours_diff)
        print("Minutes:", minutes_diff)
    except Exception as e:
        print("An error occurred:", str(e))

if __name__ == "__main__":
    main()

```

```python
# Q9. Write a Python program that reads a CSV file containing categorical data and converts a specified
# column to a categorical data type. The program should prompt the user to enter the file path, column
# name, and category order, and then display the sorted data.
# import pandas as pd
#  import numpy as np
#  columns = ['column_1', 'column_2', 'column_3', 'column_4', 'column_5', 'column_6']
#  indices = [1,2,3,4,5,6]
#  #Creating a dataframe:
#  df1 = pd.DataFrame(np.random.rand(6,6), columns = columns, index = indices)
# Answer :-
import pandas as pd

def convert_and_sort_categorical(file_path, column_name, category_order):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_path)

    # Convert the specified column to a categorical data type with the given category order
    df[column_name] = pd.Categorical(df[column_name], categories=category_order, ordered=True)

    # Sort the DataFrame based on the categorical order
    sorted_df = df.sort_values(by=column_name)

    return sorted_df

def main():
    try:
        file_path = input("Enter the file path of the CSV file: ")
        column_name = input("Enter the column name to convert to categorical: ")
        category_order = input("Enter the category order (comma-separated): ").split(',')

        sorted_df = convert_and_sort_categorical(file_path, column_name, category_order)

        print("\nSorted Data:")
        print(sorted_df)
    except FileNotFoundError:
        print("File not found. Please provide a valid file path.")
    except Exception as e:
        print("An error occurred:", str(e))

if __name__ == "__main__":
    main()

```

```python
# Q10. Write a Python program that reads a CSV file containing sales data for different products and
# visualizes the data using a stacked bar chart to show the sales of each product category over time. The
# program should prompt the user to enter the file path and display the chart.
# import pandas as pd
#  import numpy as np
#  columns = ['column_1', 'column_2', 'column_3', 'column_4', 'column_5', 'column_6']
#  indices = [1,2,3,4,5,6]
#  #Creating a dataframe:
#  df1 = pd.DataFrame(np.random.rand(6,6), columns = columns, index = indices)
# Answer :-
import pandas as pd
import matplotlib.pyplot as plt

def visualize_sales_data(file_path):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_path)

    # Create a stacked bar chart
    df.plot(x='Time', kind='bar', stacked=True, figsize=(10, 6))
    plt.title('Sales Data by Product Category over Time')
    plt.xlabel('Time')
    plt.ylabel('Sales')
    plt.legend(title='Product Category', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.xticks(rotation=45)
    plt.tight_layout()

    # Display the chart
    plt.show()

def main():
    try:
        file_path = input("Enter the file path of the CSV file: ")
        visualize_sales_data(file_path)
    except FileNotFoundError:
        print("File not found. Please provide a valid file path.")
    except Exception as e:
        print("An error occurred:", str(e))

if __name__ == "__main__":
    main()

```

```python
# Q11. You are given a CSV file containing student data that includes the student ID and their test score. Write
# a Python program that reads the CSV file, calculates the mean, median, and mode of the test scores, and
# displays the results in a table.

# The program should do the followingM
# I Prompt the user to enter the file path of the CSV file containing the student dataR
# I Read the CSV file into a Pandas DataFrameR
# I Calculate the mean, median, and mode of the test scores using Pandas toolsR
# I Display the mean, median, and mode in a table.
# Assume the CSV file contains the following columnsM
# I Student ID: The ID of the studentR
# I Test Score: The score of the student's test.
# Example usage of the program:
# Enter the file path of the CSV file containing the student data: student_data.csv
# +-----------+--------+
# | Statistic | Value |
# +-----------+--------+
# | Mean | 79.6 |
# | Median | 82 |
# | Mode | 85, 90 |
# +-----------+--------+
# Assume that the CSV file student_data.csv contains the following data:
# Student ID,Test Score
# 1,85
# 2,90
# 3,80
# 4,75
# 5,85
# 6,82
# 7,78
# 8,85
# 9,90
# 10,85

# The program should calculate the mean, median, and mode of the test scores and display the results
# in a table.
# Answer :-
import pandas as pd
from statistics import mode

def calculate_statistics(df):
    mean_score = df['Test Score'].mean()
    median_score = df['Test Score'].median()
    mode_score = mode(df['Test Score'])

    return mean_score, median_score, mode_score

def display_results(mean_score, median_score, mode_score):
    result_table = pd.DataFrame({
        'Statistic': ['Mean', 'Median', 'Mode'],
        'Value': [mean_score, median_score, mode_score]
    })

    print(result_table.to_string(index=False))

def main():
    file_path = input("Enter the file path of the CSV file containing the student data: ")
    try:
        df = pd.read_csv(file_path)
        mean_score, median_score, mode_score = calculate_statistics(df)
        print("\n+-----------+--------+")
        print("| Statistic | Value |")
        print("+-----------+--------+")
        display_results(mean_score, median_score, mode_score)
        print("+-----------+--------+")
    except FileNotFoundError:
        print("File not found. Please provide a valid file path.")
    except Exception as e:
        print("An error occurred:", str(e))

if __name__ == "__main__":
    main()
```
